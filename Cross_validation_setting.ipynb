## Primero leemos el corpus que de antemano generamos:

```python
import pandas as pd
df = pd.read_csv('/Users/user/Tesis/escritorio_anterior/MICAI_2/negativos_2/+bigrama_neg_2.csv',
                     header=0, sep=',', names=['id', 'content', 'label'])
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf_vect = TfidfVectorizer(use_idf=True, smooth_idf=True,
                             sublinear_tf=False, ngram_range=(2,2))

X = tfidf_vect.fit_transform(df['content'].values)
y = df['label'].values
```

Ahora haremos la validacion cruzada con el 33% para prueba y el 70% para entrenar en cada fold. A 10 folds, imprimimos los indices de las opiniones:

```python
from sklearn import cross_validation

#Usamos una estrategia Cross validation ShuffleSplit:
#(http://scikit-learn.org/stable/modules/generated/sklearn.cross_validation.ShuffleSplit.html
#sklearn.cross_validation.ShuffleSplit)
ss = cross_validation.ShuffleSplit(n = len(y), n_iter=10, test_size=0.30,random_state=0)


import numpy
numpy.set_printoptions(threshold=numpy.nan)

for train_index, test_index in ss:
    print "\n \n Entrenamiento 30% de 2598:\n", train_index, \
    "\n \n Prueba 70% de 2598:\n", test_index
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
```
